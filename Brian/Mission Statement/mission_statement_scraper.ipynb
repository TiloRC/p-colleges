```python

import pandas as pd
import csv
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('Data_2-14-2023.csv')

# some cleaning up (removes unnecessary columns)
df = df.drop('UnitID', axis=1)
df = df.drop('Unnamed: 4', axis=1)
df.head()
```

```python
def replace_words(p_text):
    p_text = p_text.replace('�', " ")
    p_text = p_text.replace(',', "")
    p_text = p_text.replace('\n', " ")
    p_text = p_text.replace('\t', " ")
    p_text = p_text.replace("'", "")
    p_text = p_text.replace("’", "")
    p_text = p_text.replace("’", "")
    p_list = p_text.split()
    p_text = " ".join(p_list)
    return p_text
```

```python

def scrape_mission_statement(url):
    try:
        if not url.startswith("http://") and not url.startswith("https://"):
            url = "https://" + url
        # Send a GET request to the URL
        response = requests.get(url)
        
        # Parse the HTML using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find the first occurrence of "mission statement"
        mission_heading = soup.find(lambda tag: tag.name in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6'] and 'mission' in tag.text.lower())
        print("the heading is", mission_heading.text)
        if mission_heading:
            p_tags = mission_heading.find_all_next('p')
            for p_tag in p_tags:
                if len(p_tag.text.split()) >= 6:
                    mission_statement = replace_words(p_tag.text)
                    return mission_statement

            div_tags = mission_heading.find_all_next('div')
            for div_tag in div_tags:
                if len(div_tag.text.split()) >= 6:
                    mission_statement = replace_words(div_tag.text)
                    return mission_statement

            span_tags = mission_heading.find_all_next('span')
            for span_tag in span_tags:
                if len(span_tag.text.split()) >= 6:
                    mission_statement = replace_words(span_tag.text)
                    return mission_statement
            

        mission_heading = soup.find(lambda tag: tag.name in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6'] and 'purpose' in tag.text.lower())
        if mission_heading:
            p_tags = mission_heading.find_all_next('p')
            for p_tag in p_tags:
                if len(p_tag.text.split()) >= 6:
                    mission_statement = replace_words(p_tag.text)
                    return mission_statement

            div_tags = mission_heading.find_all_next('div')
            for div_tag in div_tags:
                if len(div_tag.text.split()) >= 6:
                    mission_statement = replace_words(div_tag.text)
                    return mission_statement

            span_tags = mission_heading.find_all_next('span')
            for span_tag in span_tags:
                if len(span_tag.text.split()) >= 6:
                    mission_statement = replace_words(span_tag.text)
                    return mission_statement
                
        # test
        mission_heading = soup.find(lambda tag: tag.name in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6'] and 'value' in tag.text.lower())
        if mission_heading:
            p_tags = mission_heading.find_all_next('p')
            for p_tag in p_tags:
                if len(p_tag.text.split()) >= 6:
                    mission_statement = replace_words(p_tag.text)
                    return mission_statement

            div_tags = mission_heading.find_all_next('div')
            for div_tag in div_tags:
                if len(div_tag.text.split()) >= 6:
                    mission_statement = replace_words(div_tag.text)
                    return mission_statement

            span_tags = mission_heading.find_all_next('span')
            for span_tag in span_tags:
                if len(span_tag.text.split()) >= 6:
                    mission_statement = replace_words(span_tag.text)
                    return mission_statement

        else:
            return "CANNOT FIND STATEMENT"

    except:
        return "TOO MANY REQUEST"
```

```python
# this will take awhile
with open('output.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Institution Name', 'Mission statement (IC2020mission)']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()  # write the header row

    for row in df.values:
        insert_row = {}
        if str(row[1]) == 'nan' and not str(row[2]) == 'nan':
            insert_row['Institution Name'] = row[0]
            insert_row['Mission statement (IC2020mission)'] = scrape_mission_statement(row[2])
            # insert_row['Mission statement (IC2020mission)'] = get_mission(row[2], drop_unneeded_text(row[0]))
        elif not str(row[1]) == 'nan':
            insert_row['Institution Name'] = row[0]
            insert_row['Mission statement (IC2020mission)'] = row[1]
        else:
            print("delete", row)
        # print(insert_row)
        try:
            writer.writerow(insert_row)
        except:
            insert_row['Mission statement (IC2020mission)'] = 'nan'
            writer.writerow()
        
    
```

```python
def get_mission(url, word_list):
    try:
        if not url.startswith("http://") and not url.startswith("https://"):
            url = "https://" + url

        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        # mission statements usually contain the name of the college.
        # we will return the first paragraph instance where the name is mentioned
        for p_tag in soup.find_all('p'):
            p_text = p_tag.get_text(strip=True)
            # if (word.lower() in p_text.lower() for word in word_list):
            print("word list is", word_list)
            for word in word_list:
                if word.lower() in p_text.lower() and len(p_text.split()) > 6:
                    p_text = p_text.replace('�', " ")
                    p_text = p_text.replace(',', " ")
                    p_text = p_text.replace('\n', " ")
                    p_text = p_text.replace('\t', " ")
                    p_text = p_text.replace("'", " ")
                    return p_text
        return "CANNOT FIND MISSION STATEMENT"
    except:
        print("too many requests")
        return "TOO MANY REQUESTS"
```

```python
def drop_unneeded_text(name):
    unneeded_words = ['university', 'college', 'christian', 'institute', 'of', 'and',
                      'state', '-', 'the', 'at', 'a', 'm', '&', 'at', 'international', 'academy', 'g.', 'polytechnic', 'in', 'main', 'campus']
    words = name.lower().split()
    words_filtered = [word.replace(
        "-", " ") for word in words if word not in map(str.lower, unneeded_words)]

    list_of_words = []
    for word in words_filtered:
        for indiv_word in word.split():
            list_of_words.append(indiv_word)
    return list_of_words
```

```python
# this will take awhile
with open('output.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Institution Name', 'Mission statement (IC2020mission)']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()  # write the header row

    for row in df.values:
        insert_row = {}
        if str(row[1]) == 'nan' and not str(row[2]) == 'nan':
            insert_row['Institution Name'] = row[0]
            insert_row['Mission statement (IC2020mission)'] = get_mission(row[2], drop_unneeded_text(row[0]))
        elif not str(row[1]) == 'nan':
            insert_row['Institution Name'] = row[0]
            insert_row['Mission statement (IC2020mission)'] = row[1]
        else:
            print("delete", row)
        # print(insert_row)
        try:
            writer.writerow(insert_row)
        except:
            insert_row['Mission statement (IC2020mission)'] = 'nan'
            writer.writerow()
        
    
```
